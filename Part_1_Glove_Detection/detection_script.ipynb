```python
import os
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt

# Correct the model path with raw string or double backslashes and include weights file
model = YOLO(r"runs\detect\glove_detection3/weights/best.pt")

test_images_dir = "test/images"  # Your test images folder
output_dir = "output"                # Folder to save annotated images
os.makedirs(output_dir, exist_ok=True)

for img_file in os.listdir(test_images_dir):
    if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(test_images_dir, img_file)

        # Run inference on image
        results = model(img_path)

        # Annotate image with detections
        annotated_img = results[0].plot()

        # Save annotated image to output folder
        cv2.imwrite(os.path.join(output_dir, img_file), annotated_img)

        # Optional: Show inline in notebook (uncomment the next lines if needed)
        # img_rgb = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)
        # plt.imshow(img_rgb)
        # plt.title(img_file)
        # plt.axis('off')
        # plt.show()
```

```python
import json

output_dir = "output"  # folder for annotated images
log_dir = "logs"   # folder to save JSON logs
os.makedirs(log_dir, exist_ok=True)

for img_file in os.listdir(test_images_dir):
    if img_file.lower().endswith(('.jpg', '.jpeg', '.png')):
        img_path = os.path.join(test_images_dir, img_file)
        results = model(img_path)
        annotated_img = results[0].plot()
        cv2.imwrite(os.path.join(output_dir, img_file), annotated_img)
        
        detections = []
        boxes = results[0].boxes
        if boxes is not None and len(boxes) > 0:
            for box, label, score in zip(boxes.xyxy.cpu().numpy(), boxes.cls.cpu().numpy(), boxes.conf.cpu().numpy()):
                detections.append({
                    "label": model.names[int(label)],       # class name e.g. 'gloved_hand'
                    "confidence": float(score),             # confidence score
                    "bbox": [int(coord) for coord in box]  # bbox coordinates [x1, y1, x2, y2]
                })
        
        log_data = {
            "filename": img_file,
            "detections": detections
        }
        
        # Save one JSON log per image
        json_path = os.path.join(log_dir, img_file.replace(".jpg", ".json"))
        with open(json_path, "w") as f:
            json.dump(log_data, f, indent=2)
```
